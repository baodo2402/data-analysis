```python
df = pd.read_csv('1. Weather Data.csv')
df.head()
```

```python
#cleaning
# 1/ find all the unique 'Wind Speed' values in the data.
df['Wind Speed_km/h'].unique()
```

```python
df['Wind Speed_km/h'].nunique()
```

```python
# 2/ find the number of times when the 'Weather is exactly Clear'.
#value_counts()
df['Weather'].value_counts() # --> clear: 1326
```

```python
#filtering
df[df['Weather'] == 'Clear']
```

```python
#groupby
df.groupby('Weather').get_group('Clear')
```

```python
# 3/ find the number of times when the 'Wind Speed was exactly 4km/h'
#value_counts
df['Wind Speed_km/h'].value_counts() # --> 4 is 474
```

```python
#filtering
df[df['Wind Speed_km/h'] == 4]
```

```python
# groupby
df.groupby('Wind Speed_km/h').get_group(4)
```
